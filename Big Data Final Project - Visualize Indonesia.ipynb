```python
import pandas as pd
import numpy as np
import geopandas as gpd
import matplotlib
import matplotlib.pyplot as plt
## Load Germany Map Data
indonesia = gpd.read_file("https://raw.githubusercontent.com/bimaputra1/School_Partitipation_Rates_with_GeoPandas/master/gadm36_IDN_1.json") 
## Load Query Result
query_data = pd.read_csv("/content/drive/MyDrive/Kuliah/Semester 6/Big Data/Week 16 - UAS/data source/bq-results-percentage-update.csv")
## Merge Query result to the map data
df_merge = pd.merge(indonesia, query_data, how='left', left_on="NAME_1", right_on="sub_region_1")
## Draw MAP
fig, ax = plt.subplots(1, 1, figsize=(12,12))
df_merge.plot(ax=ax,
              column='parks_percent_change_from_baseline',
              cmap='Blues',
              scheme='quantiles',
              legend=True,
              linewidth=0.8,);
```
